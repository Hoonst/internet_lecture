```python
import pandas as pd
import sys
import re
```

```python
file=open('text.txt','r')

lines=file.readlines()
```

```python
lines
```

```python
lines = [x for x in lines if x != '\n']
```

```python
lines
```

```python
def isHangul(text):
    #Check the Python Version
    pyVer3 =  sys.version_info >= (3, 0)

    if pyVer3 : # for Ver 3 or later
        encText = text
    else: # for Ver 2.x
        if type(text) is not unicode:
            encText = text.decode('utf-8')
        else:
            encText = text

    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))
    return hanCount > 0
```

```python
english_only = []
```

```python
for line in lines:
    if '①\n' in line or '②\n' in line or '③\n' in line or '④\n' in line or '⑤\n' in line:
        pass
    
    elif isHangul(line) == False:
        english_only.append(line)
        
    
        
```

```python
one_for_all = '.'.join(english_only)
```
