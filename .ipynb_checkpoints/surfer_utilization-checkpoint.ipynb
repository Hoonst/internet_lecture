```python
def set_driver():
    global driver, actionChains
    driver = webdriver.Chrome()
    actionChains = ActionChains(driver)
    
def connect_site_with_url(site_url, time):
    driver.get(site_url)
    implicit_wait(time)
    
def username_password_insert(username, password):
    user = driver.find_element_by_id("username")
    pw = driver.find_element_by_id("password")
    user.send_keys(username)
    pw.send_keys(password)

    driver.find_element_by_id("signin_button").click()
    
def implicit_wait(time):
    driver.implicitly_wait(time)

def scroll_very_bottom(very_bottom_element, time):
    driver.execute_script('arguments[0].scrollIntoView(true);', very_bottom_element)
    implicit_wait(time)
    
def actual_performance(plausible_name):
    driver.get(division_dic['division_1'])
    implicit_wait(5)
    try:
        to_element = driver.find_element_by_css_selector("upload-modal.ng-scope")
        scroll_very_bottom(to_element, 5)
        implicit_wait(5)

        if driver.find_elements_by_xpath(f"//*[contains(text(), '{plausible_name}')]") != []:
#           print(driver.find_elements_by_xpath(f"//*[contains(text(), '{plausible_name}')]"))
            actionChains = ActionChains(driver)
            button = driver.find_elements_by_xpath(f"//*[contains(text(), '{plausible_name}')]")[0]
            actionChains.move_to_element(button).context_click(button).perform()

            implicit_wait(5)

            content = driver.find_element_by_css_selector('td.link > a.ng-binding')
            text = content.text
            
            #Some-how updating cell function's index doesn't start from 0
            #test_sheet.update_cell(row, 12, text)
            
    except Exception as e:
        print(e)
        
def download_image_with_url(url, name):   
    os.system(f'wget --output-document={name}{Path(url).suffix} {url}')
    
    def determine_s3_component(url):
        logo_url_prefix_ver1 = 'https://s3.amazonaws.com/footballpedia/logo/'
        logo_url_prefix_ver2 = 'https://footballpedia.s3.amazonaws.com/logo/'
        if (logo_url_prefix_ver1 in url) or (logo_url_prefix_ver2 in url):
            return True
        else:
            return False
```

```python
max_number = find_max_page()

# hoonk_list = driver.find_elements_by_tag_name('tr')
def main_function(max_number, index_from = False, index_to = False):
    
    for index in range(1, max_number):
        switch_frame()

        ###
        hoonk_list = []
        for hoon in driver.find_elements_by_tag_name('tr'):
            if '윤훈관 선생님' in hoon.text:
                hoonk_list.append(hoon)

        for i in range(len(hoonk_list)):
            switch_frame()
            hoonk_list = []

            for hoon_2 in driver.find_elements_by_tag_name('tr'):
                if '윤훈관 선생님' in hoon_2.text:
                    hoonk_list.append(hoon_2)

            #선생 답변의 제목 추출 및 클릭
            switch_frame()
            index_number = hoonk_list[i].find_element_by_tag_name('td').text
            
            if index_from != False or index_to != False:
                if int(index_number) in range(300):
                    print('This is where we draw the line')
                    return ''

            a = hoonk_list[i].find_element_by_css_selector('td > a')

            teacher_answer_title = a.text
            a.click()

            switch_frame()


            #--------------------------------------------------------------------------------
            # 학생 이름/날짜/정보/출처
            name = driver.find_elements_by_tag_name('td')[7].text
            date = driver.find_elements_by_tag_name('td')[9].text
            info = driver.find_elements_by_tag_name('td')[13].text.split('\n')
            clas = info[0]
            lecture = info[1]
            origin = driver.find_elements_by_tag_name('td')[17].text

            #학생 질문 제목 / 내용/ 선생 답변 내용 추출 
            question_title = driver.find_elements_by_tag_name('td')[3].text[:-13]
            driver.switch_to.frame('Qa_cont')
            # 학생 질문 내용
            #print(driver.find_element_by_tag_name('table').text.replace('\n', ''))
            question_content = driver.find_element_by_tag_name('table').text.replace('\n', '')
            #--------------------------------------------------------------------------------

            #--------------------------------------------------------------------------------
            # 선생 답변 내용
            switch_frame()

            driver.switch_to.frame('ifrESPRESSO_EDITOR')
            driver.switch_to.frame(driver.find_element_by_id('espresso_WYSIWYGEditor_WYSIWYGEditor'))

            #print(driver.find_element_by_tag_name('body').text.replace('\n', ''))
            teacher_answer_content = driver.find_element_by_tag_name('body').text.replace('\n', '')
            #--------------------------------------------------------------------------------

            in_total = {'id': index_number,
                        '작성자':name,
                        '작성일':date,
                        '강좌명':clas,
                        '강의명':lecture,
                        '질문 채널':origin,
                        '질문 제목':question_title, 
                        '질문 내용':question_content, 
                        '답변 제목':teacher_answer_title, 
                        '답변 내용':teacher_answer_content}

            df = df.append(in_total, ignore_index=True)
            # 돌아가기
            driver.back()



        # select next_page
        switch_frame()
        scroll = driver.find_elements_by_tag_name('table')[-1]
        scroll.find_element_by_link_text(str(index+1)).click()
```

# Connecting to gspread

```python
def connect_to_sheet(sheet_name):
    scope = ['https://spreadsheets.google.com/feeds',  'https://www.googleapis.com/auth/drive']
    creds = ServiceAccountCredentials.from_json_keyfile_name('client_secret.json', scope)
    client = gspread.authorize(creds)
    
    sheet = client.open(sheet_name).sheet1

    # Extract and print all of the values
    list_of_hashes = sheet.get_all_records()
    #print(list_of_hashes)
    return sheet

test_sheet = connect_to_sheet("SportsPedia NCAA Football Team Links")
```

```python
logo_url_prefix_ver1 = 'https://s3.amazonaws.com/footballpedia/logo/'
logo_url_prefix_ver2 = 'https://footballpedia.s3.amazonaws.com/logo/'
for row in range(495, 768):
    driver.get('https://s3.console.aws.amazon.com/s3/buckets/footballpedia/logo/?region=ap-northeast-2&tab=overview')
    logo_sheet_row = test_sheet.row_values(row)
    division = logo_sheet_row[3]
    school = '_'.join(logo_sheet_row[4].split(' '))
    
    nickname = '_'.join(logo_sheet_row[7].split(' '))
    plausible_name = f'{school}_{nickname}'
    
    try:
        logo_url = logo_sheet_row[11]
    except Exception as e:
        print(e)
    print(row, division, plausible_name)
    print(logo_url)
    

    if determine_s3_component(logo_url):
        continue
    
    elif division == 'FBS' or division == 'FCS':
        connect_site_with_url(division_dic['division_1'], 5)
        actual_performance(plausible_name)
        
    elif division == 'D2':
        connect_site_with_url(division_dic['division_2'], 5)
        actual_performance(plausible_name)
        
    elif division == 'D3':
        connect_site_with_url(division_dic['division_3'], 5)
        actual_performance(plausible_name)
        
    elif division == 'NAIA':
        connect_site_with_url(division_dic['division_naia'], 5)
        actual_performance(plausible_name)
```
