```python
dic = defaultdict(int)
keyword, page = '윤훈관', 1
b.get('https://m.cafe.naver.com/ArticleSearchList.nhn?search.clubid=10197921&search.searchdate=all&search.searchBy=0&search.query='+keyword+'&search.defaultValue=1&search.includeAll=&search.exclude=&search.include=&search.exact=&search.sortBy=date&userDisplay=15&search.media=0&search.option=0&search.page='+str(page))


page_number = b.find_element_by_css_selector('span.cnt_wrap').text
page_number = page_number.split()
total_page = int(page_number[1]) // int(page_number[0])

while page <= total_page:
    print('page {}'.format(page))
    b.get('https://m.cafe.naver.com/ArticleSearchList.nhn?search.clubid=10197921&search.searchdate=all&search.searchBy=0&search.query='+keyword+'&search.defaultValue=1&search.includeAll=&search.exclude=&search.include=&search.exact=&search.sortBy=date&userDisplay=15&search.media=0&search.option=0&search.page='+str(page))
    rows = b.find_elements_by_css_selector('ul.list_tit > li')
    for row in rows:
        time = row.find_element_by_css_selector('span.time')
        print(time.text)
        today = str(datetime.date.today())

        if ':' in time.text:
            dic[today] += 1
        else:
            dic[str(datetime.datetime.strptime(time.text, '%y.%m.%d.').date())] += 1
        
    page += 1
    
    
    
```

```python
teacher, page = '이명학', 1
orbi_dic = defaultdict(int)
while True:
    b.get("https://orbi.kr/search?q={}&type=keyword&page={}".format(str(teacher), str(page)))
    rows = b.find_elements_by_css_selector("ul.post-list > li")
    print('page {}'.format(str(page)))
    if rows == []:
        break
    for row in rows:
        date = row.find_element_by_tag_name('abbr').text
        date = date.split()[0]
        if '/' not in date: 
            orbi_dic[str(datetime.date.today())] += 1
        else:
            if date == '02/29':
                orbi_dic['2020-02-29'] += 1
            
            elif len(date) < 6:
                date = datetime.datetime.strptime(date, '%m/%d').date()
                date = date.replace(year=2020)
                print(date)
                orbi_dic[str(date)] += 1
            
            else:
                date = datetime.datetime.strptime(date, '%y/%m/%d').date()
                print(date)
                orbi_dic[str(date)] += 1
    page += 1
```
